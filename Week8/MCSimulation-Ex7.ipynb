```python
newmask=(samples_y<f(samples_x))plot(x,f(x),lw=3)plot(samples_x[1:200],samples_y[1:200], '.')plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')plot(samples_x[newmask[:200]], samples_y[newmask[:200]], 'r.')
```
