```python
f = lambda x: x**(0.5)+sin(x)
```

```python
x=arange(0.,15.,0.001)
```

```python
xmin=1.7xmax=12.4#find the max val of f within the bound set ymaxsubx=x[logical_and((x>xmin),(x<xmax))]ymax=max(f(subx))*1.05ymin=0.#replotmyplot=plot(x,f(x),lw=3)xlim=(0.,15.)#plot the boxplot([xmin,xmin],[ymin,ymax],color='k',linestyle='--')#vertical leftplot([xmax,xmax],[ymin,ymax],color='k',linestyle='--')#vertical rightplot([xmin,xmax],[ymax,ymax],color='k',linestyle='--')#horizontal top
```
