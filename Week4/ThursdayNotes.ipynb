Using boolean masks to find data--------------------------------

```python
theta = arange(0.,2*pi,pi/100)plot(theta, sin(theta))show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)#print maskplot(theta,sin(theta),theta[mask],sin(theta[mask]))
```

```python
y = sin(theta[mask])jj = ( y == max(y))y[jj]=NaN#Not a Number, computer way of saying infiinity, computer knows to ignore/git rid of
```

```python
plot(theta,sin(theta))plot(theta[mask],y,linewidth=4)show()
```

Interpolation-------------_____________

```python
from scipy.interpolate import interp1d
```

```python
theta=arange(0.,2*pi+.01,pi/10)mask = (theta==0.5)print mask
```

```python
newtheta=0.5; newsin=0, sininterp = interp1d(theta,sin(theta))#help(interp1d)#created interpolating function, now using it to calc sin at new thetanewsin= sininterp(newtheta)print newsin
```

```python
plot(theta,sin(theta),newtheta,newsin, 'm+',markersize=20,markeredgewidth=3)
```

```python
thetanew=arange(0,2*pi,pi/30)#adding the extra () at the end makes everything before it act like a functionsinnew=interp1d(theta,sin(theta))(thetanew)plot(theta,sin(theta), thetanew, sinnew, 'r*')show()
```

```python
plot(thetanew, sinnew-sin(thetanew))ylabel='Make a label'show()
```

```python
sinnewc=interp1d(theta,sin(theta),kind='cubic')(thetanew)subplot(2,1,1) plot(thetanew,sinnew-sin(thetanew),thetanew,sinnewc-sin(thetanew), 'r^')subplot(2,1,2)plot(thetanew,sinnewc-sin(thetanew), 'r')
```
