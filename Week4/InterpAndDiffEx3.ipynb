3. Let’s explore the accuracy of numerical differentiation.

(a) Write a function called finiteDifference(x,y) that takes as arguments arrays ofvalues x and y(x) and returns the derivative of y with respect to x. Use the centered finitedifference method so that the dydx returned by the function has the same dimensions asthe x,y arrays.

```python
def finiteDiff(x,y):      """returns the derivative of y WRT x"""      dydx = zeros(y.shape,float)      dydx[1:-1] = (y[2:]-y[:-2])/(x[2:]-x[:-2])      dydx[0] = (y[1]-y[0])/(x[1]-x[0])      dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])      return dydx
```
